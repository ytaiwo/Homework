```python
from bs4 import BeautifulSoup as bs
import requests

#need to look at terms and conditions to know if you can scrape the website.
#most websites have a file called robots.txt that you can look at to see if scraping is allowed
url = "http://quotes.toscrape.com/"
```

```python
#get page with requests
response = requests.get(url)
response
```

```python
soup = bs(response.text,'html.parser')
print(soup.prettify())
```

```python
print(soup.title)
```

```python
all_divs = soup.find_all('div',class_='quote')
all_divs
```

```python
import pandas as pd
page_one_results = []
for div in all_divs:
    try:
        
        quote_text = div.find('span',class_='text').text
        author = div.find('small',class_='author').text
        quote_dict = {'quote':quote_text,
                        'author': author}
        page_one_results.append(quote_dict)
    except AttributeError as e:
        print(e)
page_one_df = pd.DataFrame(page_one_results)
page_one_df
```

```python
all_quotes = []

for i in range(1,11):
    url = f'https://quotes.toscrape.com/page/{i}'
    
    for div in all_divs:
        try:
            quote_text = div.find('span',class_='text').text
            author = div.find('small',class_='author').text
            tags = div.find('div',class_='tags').find_all('a')
            tags_list = []
            for tag in tags:
                tags_list.append(tag.text)
                
            quote_dict = {'quote':quote_text,
                          'author': author,
                          'tag':tags_list}
            
            all_quotes.append(quote_dict)
        except AttributeError as e:
            print(e)
        
df = pd.DataFrame(all_quotes)
df
```

## chrome driver

```python
from splinter import Browser 
from webdriver_manager.chrome import ChromeDriverManager

executable_path = {'executable_path':ChromeDriverManager().install()}
browser = Browser('chrome',**executable_path,headless=False)
```

```python
all_quotes = []
url = 'https://quotes.toscrape.com/'
browser.visit(url)

for i in range(1,5):
    soup = bs(browser.html,'html.parser')
    all_divs = soup.find_all('div',class_='quote')
    for div in all_divs:
        try:
            quote_text = div.find('span',class_='text').text
            author = div.find('small',class_='author').text
            tags = div.find('div',class_='tags').find_all('a')
            tags_list = []
            for tag in tags:
                tags_list.append(tag.text)
                
            quote_dict = {'quote':quote_text,
                          'author': author,
                          'tag':tags_list}
            
            all_quotes.append(quote_dict)
            
        except AttributeError as e:
            print(e)
    browser.links.find_by_partial_text('Next').click()
df = pd.DataFrame(all_quotes)
df
```
