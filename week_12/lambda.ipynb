# Lambda Functions

A lambda function is an anonymous function, meaning it doesnt have a name. So, it doesnt start with def and it has less code. We can use lambdas as functions, to filter results, apply results to a column, call a more complicated function, etc. 

```python
import pandas as pd
#"C:\Users\alm0708\Documents\Data-Science-Assignments\course_material\week_05\JEOPARDY_CSV.csv"
jeopardy_df = pd.read_csv("../week_05/JEOPARDY_CSV.csv")
jeopardy_df.head()
```

```python
jeopardy_df.columns = jeopardy_df.columns.str.replace(" ","")
jeopardy_df.head()
```

```python
jeopardy_df.dtypes
```

```python
jeopardy_df['AirDate'] = pd.to_datetime(jeopardy_df['AirDate'],format='%m/%d/%Y')
jeopardy_df.dtypes
```

```python
jeopardy_df = jeopardy_df.assign(month=lambda x: x['AirDate'].dt.month)
jeopardy_df.head()
```

```python
#how do we do the same thing using apply?
jeopardy_df['month2']= jeopardy_df.apply(lambda x: x['AirDate'].month, axis = 1)
jeopardy_df.head()
```

# Filter

```python
value_list = jeopardy_df['Value'].tolist()
value_list=value_list[0:50]
value_list
```

```python
filtered_list = list(filter(lambda num:int(num.replace('$','').replace(' ','').replace(',',''))>300,value_list))
filtered_list[0:10]
```

# Map and Reduce
Map applies the logic to a whole list. map(lambda, list) can be applied to a whole dataframe or column or list.

Reduce performs an aggregation or calculation on a whole list. 


```python
filtered_list = list(map(lambda num:int(num.replace('$','').replace(' ','').replace(',',''))>300,value_list))
filtered_list[0:10]
```
