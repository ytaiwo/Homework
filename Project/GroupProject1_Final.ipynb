## Univariate Analysis:

#### Plotting Histogram to check the skewness of Labour Force Participation - Female 15+ yr (Percent)

```python
plt.hist(UN_df['Labour Force Participation - Female 15+ yr (Percent)'],bins='auto')
plt.axvline(UN_df['Labour Force Participation - Female 15+ yr (Percent)'].mean(),color='green',label='Mean')   
plt.axvline(UN_df['Labour Force Participation - Female 15+ yr (Percent)'].median(),color='blue',label='Median')
plt.axvline(UN_df['Labour Force Participation - Female 15+ yr (Percent)'].mode()[0],color='red',label='Mode1')
plt.axvline(UN_df['Labour Force Participation - Female 15+ yr (Percent)'].mode()[1],color='red',label='Mode2')
plt.axvline(UN_df['Labour Force Participation - Female 15+ yr (Percent)'].mode()[2],color='red',label='Mode3')
plt.xlabel("Labour Force Participation - Female 15+ yr (Percent)")
plt.ylabel("Frequency")
plt.title("Female Participation Spread")
plt.savefig('female_labourforce_histogram.png')
```

#### Data for the 'Labour Force Participation - Female 15+ yr (Percent)' is plotted on the histogram.
#### 1.Mean is at 58.
#### 2.Median is between 58 and 59. Much closer to 59.
#### 3.Plotted all the 3 modes and all three are towards the right side of the mode.
#### 4.Hence data is left skewed.
#### 5.Central tendency for the data is median.

#### Plotting Histogram to check the skewness of Labour Force Participation - Male 15+ yr (Percent)

### Question 3: What is the comparison between labour force participation for male and female for each age group?

```python
fig, ax = plt.subplots(5,1,sharex=True,figsize=(7,7))

ax[0].plot(UN_df.index, UN_df['Labour Force Participation - Female 15-24 yr (Percent)'], color='b', label = 'Female 15-24 yr')
ax[0].plot(UN_df.index, UN_df['Labour Force Participation - Male 15-24 yr (Percent)'], color='r', label = 'Male 15-24 yr')

ax[1].plot(UN_df.index, UN_df['Labour Force Participation - Female 25-34 yr (Percent)'], color='g', label = 'Female 25-34 yr')
ax[1].plot(UN_df.index, UN_df['Labour Force Participation - Male 25-34 yr (Percent)'], color ='orange' , label = 'Male 25-34 yr')

ax[2].plot(UN_df.index, UN_df['Labour Force Participation - Female 35-54 yr (Percent)'], color='purple' , label = 'Female 35-54 yr')
ax[2].plot(UN_df.index, UN_df['Labour Force Participation - Male 35-54 yr (Percent)'],color = 'y' , label = 'Male 35-54 yr')

ax[3].plot(UN_df.index, UN_df['Labour Force Participation - Female 55-64 yr (Percent)'], color = 'c' ,label = 'Female 55-64 yr')
ax[3].plot(UN_df.index, UN_df['Labour Force Participation - Male 55-64 yr (Percent)'], color = 'indigo', label = 'Male 55-64 yr')

ax[4].plot(UN_df.index, UN_df['Labour Force Participation - Female 65+ yr (Percent)'], color = 'k', label = 'Female 65+ yr')
ax[4].plot(UN_df.index, UN_df['Labour Force Participation - Male 65+ yr (Percent)'], color = 'pink' , label = 'Male 65+ yr')

ax[4].set_xlabel("Year")
plt.xticks(rotation=45)
ax[2].set_ylabel("Percent of Labor Force Participation")
fig.suptitle("Female vs Male Labor Force Participation")
fig.legend(loc=1, prop={'size':7})
plt.show()

plt.savefig('female_male_labourforce_subplot.png')
```

#### Analysis: 
#### 1. In 2004, there is a sharp drop in employment in the 15-24 age range in both male and female. Drop is seen less so in other age groups.
#### 2. The gap in labour force participation between male and female is getting closer to each other, except in the 65+ age range.
