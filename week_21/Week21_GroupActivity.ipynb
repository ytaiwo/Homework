```python
import requests
import pandas as pd
import numpy as np
import re
import json
```

```python
#from bs4 import BeautifulSoup
import bs4

base_url = "https://flask-app-class-demo-yt.azurewebsites.net/gdp/"
years = range(1960,2021)
gdp_list = []

for year in years:
    
    full_request = base_url + str(year)
    #print(full_request)

    req = requests.get(full_request)

    req.encoding = 'utf-8'

    content = req.text
    
    new_content = bs4.BeautifulSoup(content)
    content_string = new_content.text
    
    mod_string = re.search('{.*}', content_string).group(0)
    json_string = mod_string.replace("'",'"')
    gdp_data = json.loads(json_string)
    
    gdp_list.append([gdp_data['indicator']['id'], gdp_data['indicator']['value'], gdp_data['country']['id'],
                    gdp_data['country']['value'], gdp_data['countryiso3code'],gdp_data['date'],gdp_data['value'],
                    gdp_data['unit'],gdp_data['obs_status'],gdp_data['decimal']])
    
    
#create dataframe
gdp_df = pd.DataFrame(gdp_list,columns=['indicator_id','indicator_value','country_id','country_value','countryiso3code',
                                        'date', 'value', 'unit', 'obs_status', 'decimal'])
gdp_df
```
