```python
import pandas as pd

belize_path = "belize_airbnb_listings.csv"
belize_df = pd.read_csv(belize_path)
```

```python
belize_df.head()
```

```python
neighbourhood = belize_df.groupby("neighbourhood")

neighbourhood_info = pd.DataFrame({"Maximum Price":neighbourhood["price"].max(),
                                   "Average Price":neighbourhood["price"].mean()})

neighbourhood_info
```
