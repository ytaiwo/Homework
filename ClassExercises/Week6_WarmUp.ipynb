```python
import pandas as pd
import numpy as np

belize = pd.read_csv('belize_airbnb_listings.csv')
belize
```

```python
#group data by neighborhoods
neighbourhood = belize.groupby("neighbourhood")
avg_price = neighbourhood["price"].mean()
reviews = neighbourhood["number_of_reviews"].sum()
availability = neighbourhood["availability_365"].mean()

#Create a dataframe :
data = {
    "Average Price" :avg_price,
    "Number of Reviews" :reviews,
    "Availability 365" :availability  
}

neighbourhood_df = pd.DataFrame(data)
neighbourhood_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x ="neighbourhood", y = "Average Price", data=neighbourhood_df, hue="Average Price")
plt.xticks(rotation = 90)
plt.show()
```
