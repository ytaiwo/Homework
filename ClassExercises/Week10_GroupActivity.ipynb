```python
import numpy as np

def dotproduct(a,b):
    dot_product = 0 
    try:
        for a,b in zip(a,b):
            dot_product = dot_product + a*b
        return dot_product
    except:
        print("Sorry, the arrays should be the same size")
```

```python
a = np.array([3, 2, 2])

b = np.array([3, 6, 9])

dotproduct(a,b)
```

```python
def crossproduct(a,b):
    dim = len(a)
    cross = []
    try:
        for i in range(dim):
            if i == 0:
                j,k = 1,2
                cross.append(a[j]*b[k] - a[k]*b[j])
            elif i == 1:
                j,k = 2,0
                cross.append(a[j]*b[k] - a[k]*b[j])
            else:
                j,k = 0,1
                cross.append(a[j]*b[k] - a[k]*b[j])
        return cross
    except:
        print("Arrays should be same size")
```

```python
a = np.array([3, 0, 2])

b = np.array([3, 1, 2])

crossproduct(a,b)
```

## One Hot Encoding

### One hot encoding is representing a categorical value as binary vectors by transforming the categories to integer form. 
### Used so that python can work with categorical values since they're in integer format.
