```python
import pandas as pd

jeopardy = pd.read_csv("JEOPARDY_CSV.csv")
jeopardy.head()
```

```python
#grain is the grainularity of data - so it's the question.
#clean the data by removing extra spaces in the column titles
```

```python
jeopardy.columns = jeopardy.columns.str.rstrip()
jeopardy.columns = jeopardy.columns.str.lstrip()

jeopardy.columns
```

```python
jeopardy.describe()
#There are 216,930 rows. A row represents the show number, air date, round, category, value ,question and answer for a jeopardy game.
```

```python
jeopardy['Year'] = pd.DatetimeIndex(jeopardy['Air Date']).year
jeopardy.head()
```

```python
show_info = jeopardy.groupby('Year')

shows = show_info['Show Number'].nunique()

shows_by_year = pd.DataFrame({"Unique Shows Each Year": shows})

shows_by_year
```

```python
category_info = jeopardy["Category"].value_counts()
category_count = pd.DataFrame({"Number of Times Category Used": category_info})
category_count
```

```python
jeopardy['Month'] = pd.DatetimeIndex(jeopardy['Air Date']).month
jeopardy.head(20)
```

```python
most_shows = jeopardy["Month"].value_counts()
most_shows_df = pd.DataFrame({"Most Shows": most_shows})
most_shows_df
```

```python
common_answers = jeopardy["Answer"].value_counts().head(3)

top_answers = pd.DataFrame({"Common Answer Count": common_answers})
top_answers
```

```python
common_ques = jeopardy.loc[jeopardy['Answer'].isin(["China", "Australia", "Japan"])]
common_ques
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

jeopardy_categories=jeopardy['Category'].value_counts().to_frame()
category = jeopardy_categories.reset_index().head(5)

sns.barplot(x = "index", y="Category", data = category).set_title("Top 5 Jeopardy Categories")
plt.xticks(rotation = 90)
plt.show()
```
