```python
def descending_order(n = 11111):
    number_str = str(n)
    sort_number = ''.join(sorted(number_str, reverse = True))
    return sort_number
    
descending_order(43905)
```

```python
import pandas as pd

import_file = "JEOPARDY_CSV.csv"

jeopardy_df = pd.read_csv(import_file)
jeopardy_df.head()
```

```python
jeopardy_df.describe()
```

```python
#2a. There are 216930 rows and a row represents a jeopardy question
```

```python
jeopardy_df.columns = jeopardy_df.columns.str.rstrip()
jeopardy_df.columns = jeopardy_df.columns.str.lstrip()

jeopardy_df.columns
```

```python
jeopardy_df['Year'] = pd.DatetimeIndex(jeopardy_df['Air Date']).year
jeopardy_df.head()
```

```python
show_info = jeopardy_df.groupby('Year')

shows = show_info['Show Number'].nunique()

shows_by_year = pd.DataFrame({"Unique Shows Each Year": shows})

shows_by_year
```

```python
#Max Value of each category

category_info = jeopardy_df.groupby("Value")

categories = category_info["Category"].max()

categories_df = pd.DataFrame({"Category": categories})
categories_df
```
